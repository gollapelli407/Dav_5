```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

np.random.seed(42)
normal_patients = np.random.normal(100, 15, 150)
high_patients = np.random.normal(180, 20, 50)
blood_sugar_levels = np.concatenate([normal_patients, high_patients])

plt.figure(figsize=(8,5))
sns.kdeplot(blood_sugar_levels, fill=True, color='skyblue', linewidth=2)

plt.title("Density Plot of Blood Sugar Levels of 200 Patients", fontsize=14)
plt.xlabel("Blood Sugar Level (mg/dL)", fontsize=12)
plt.ylabel("Density", fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)

plt.show()
```

```python
# Import required libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Example COVID data
data = {
    'State': ['Maharashtra', 'Kerala', 'Delhi', 'Karnataka', 'Tamil Nadu'],
    'Cases (in thousands)': [320, 250, 150, 180, 200]
}

df = pd.DataFrame(data)
df.set_index('State', inplace=True)

# Create a heatmap
plt.figure(figsize=(6, 4))
sns.heatmap(df, annot=True, cmap='Reds', linewidths=0.5)
plt.title("COVID-19 Cases Heatmap (2020)")
plt.show()
# Import libraries
import geopandas as gpd

# Reload data for merge
data = {
    'State': ['Maharashtra', 'Kerala', 'Delhi', 'Karnataka', 'Tamil Nadu'],
    'Cases': [320, 250, 150, 180, 200]
}
df = pd.DataFrame(data)

# Load India map GeoJSON (open-source)
url = "https://raw.githubusercontent.com/geohacker/india/master/state/india_telengana.geojson"
india = gpd.read_file(url)

# Merge COVID data with map
merged = india.merge(df, left_on='st_nm', right_on='State', how='left')

# Plot geospatial map
plt.figure(figsize=(10, 8))
merged.plot(column='Cases',
            cmap='Reds',
            legend=True,
            edgecolor='black',
            linewidth=0.8)
plt.title("Geospatial Visualization of COVID-19 Cases (India, 2020)", fontsize=14)
plt.axis('off')
plt.show()
```
